```python
import numpy as np #operaciones matriciales y con vectores
import pandas as pd #tratamiento de datos
import matplotlib.pyplot as plt #gráficos
from sklearn import naive_bayes, metrics
from sklearn.model_selection import cross_val_score, cross_validate #método para evaluar varios particionamientos de C-V
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
data = pd.read_csv('data_set_cleaned.csv', header = 0, names=None)
data.head()
```

```python
x = data.loc[: , ~data.columns.isin(['Case1_Control0','Genero'])].values
y = data['Case1_Control0'].values
```

```python
gnb = naive_bayes.GaussianNB()
kf = KFold(n_splits=10, shuffle=True, random_state=1234)
acc_test_vec = []
for indices_train, indices_test in kf.split(x):
    x_train = x[indices_train]
    x_test = x[indices_test]
    modelGNB = gnb.fit(x_train, y[indices_train])
    y_pred = modelGNB.predict(x_test)
    acc_test_vec.append(metrics.accuracy_score(y[indices_test], y_pred))
#acc_test_vec
print('Exatitud promedio alcanzada por el modelo %0.4f' %(np.mean(acc_test_vec)))
```

Este modelo fue el que pero rendimiento tuvo durante el entrenamiento, por tanto es descartado.
